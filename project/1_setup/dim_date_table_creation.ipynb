**Import Required Libraries**

```python
import pandas as pd
import numpy as np
from datetime import datetime
import os
```

## Date Dimension Table Generator

This notebook creates a monthly-grain date dimension table with the following columns:

* **month_start_date** - First day of each month
* **date_key** - Surrogate key in YYYYMM format (e.g., 202401)
* **year** - Calendar year
* **month_name** - Full month name (e.g., "January")
* **month_short_name** - Abbreviated month name (e.g., "Jan")
* **quarter** - Quarter (e.g., "Q1")
* **year_quarter** - Year and quarter (e.g., "2024-Q1")

**Output:** Saved as CSV file in `data/dim_date.csv`

**Define start and end dates**

```python
# start date and end date
start_date = "2024-01-01"
end_date   = "2025-12-01"
```

```python
# 1️⃣ Generate one row per month start between start_date and end_date
date_range = pd.date_range(start=start_date, end=end_date, freq='MS')  # MS = Month Start
df = pd.DataFrame({'month_start_date': date_range})

print(f"Generated {len(df)} month records from {start_date} to {end_date}")

# 2️⃣ Add useful analytics columns
df['date_key'] = df['month_start_date'].dt.strftime('%Y%m').astype(int)
df['year'] = df['month_start_date'].dt.year
df['month_name'] = df['month_start_date'].dt.strftime('%B')  # Full month name
df['month_short_name'] = df['month_start_date'].dt.strftime('%b')  # Abbreviated month name
df['quarter'] = 'Q' + df['month_start_date'].dt.quarter.astype(str)
df['year_quarter'] = df['year'].astype(str) + '-Q' + df['month_start_date'].dt.quarter.astype(str)

print("\nAdded columns: date_key, year, month_name, month_short_name, quarter, year_quarter")
```

```python
display(df)
```

```python
# Create output directory if it doesn't exist
os.makedirs("data", exist_ok=True)

# Save as CSV file
output_path = "data/dim_date.csv"
df.to_csv(output_path, index=False)

print(f"✓ Date dimension table saved to: {output_path}")
print(f"  Total records: {len(df)}")
print(f"  Date range: {df['month_start_date'].min()} to {df['month_start_date'].max()}")
print(f"\nColumns: {list(df.columns)}")
```
